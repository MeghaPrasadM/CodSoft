```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/advertising.csv")
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
sns.regplot(x=df['TV'],y=y,color='r')
```

```python
sns.regplot(x=df['Newspaper'],y=y,color='b')
```

```python
sns.regplot(x=df['Radio'],y=y,color='g')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=9)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual value':y_test,'Predicted value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
print(model.predict([[31.3,12.3,21.4]]))
```

```python
print("Slope is")
list(zip(x,model.coef_))
```

```python
print("Constant is",model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print("MAPE is",mean_absolute_percentage_error(y_test,y_pred))
print("r2_score is",r2_score(y_test,y_pred))
```
